```python
import pandas as pd
from models import *
```

```python
assuntos = get_keywords()
print("assuntos: ", ",".join(assuntos))
```

```python
### G1
def get_g1(assunto,num=0):
    print('iniciando a busca do G1')
    try:
        g1 = pd.read_csv('dados_g1190920.csv.csv',sep='|',encoding='utf-8')
        g1_link = list(g1['link'])
    except:
        g1 = pd.DataFrame()
        g1_link = []
    print('já temos '+str(len(g1_link))+' links em G1')
    links = get_links_from_globo(assunto,i,'g1') + get_links_from_google('g1.globo.com',assunto,num)
    links = [x for x in list(set(links)) if x not in list(g1_link)]
    print('buscando em '+str(len(links))+' links em G1')
    for link in links:
        g1 = g1.append(get_news(link.replace(' ','%20'),'g1'), ignore_index=True)
    g1.to_csv('dados_g1190920.csv',index=False,sep='|',encoding='utf-8')
    return g1
```

```python
### R7
def get_r7(assunto,num=0):
    print('iniciando a busca do R7')
    try:
        r7 = pd.read_csv('dados_r7190920.csv',sep='|',encoding='utf-8')
        r7_link = list(r7['link'])
    except:
        r7 = pd.DataFrame()
        r7_link = []
    print('já temos '+str(len(r7_link))+' links em R7')
    links = get_links_from_google('noticias.r7.com',assunto,num)
    links = [x for x in list(set(links)) if x not in r7_link]
    print('buscando em '+str(len(links))+' links em R7')
    for link in links:
        r7 = r7.append(get_news(link.replace(' ','%20'),'r7'), ignore_index=True)
    r7.to_csv('dados_r7190920.csv',index=False,sep='|',encoding='utf-8')
    return r7
```

```python
### UOL
def get_uol(assunto,num=0):
    print('iniciando a busca do UOL')
    try:
        uol = pd.read_csv('dados_uol190920.csv',sep='|',encoding='utf-8')
        uol_link = list(uol['link'])
    except:
        uol = pd.DataFrame()
        uol_link = []
    print('já temos '+str(len(uol_link))+' links em UOL')
    links = get_links_from_google('noticias.uol.com.br',assunto,num)
    links = [x for x in list(set(links)) if x not in uol_link]
    print('buscando em '+str(len(links))+' links em UOL')
    for link in links:
        uol = uol.append(get_news_uol(link.replace(' ','%20')), ignore_index=True)
    uol.to_csv('dados_uol190920.csv',index=False,sep='|',encoding='utf-8')
    return uol
```

```python
### OGLOBO
def get_oglobo(assunto,num=0):
    print('iniciando a busca do OGLOBO')
    try:
        oglobo = pd.read_csv('dados_oglobo190920.csv',sep='|',encoding='utf-8')
        oglobo_link = list(oglobo['link'])
    except:
        oglobo = pd.DataFrame()
        oglobo_link = []
    print('já temos '+str(len(oglobo_link))+' links em Oglobo')
    links = get_links_from_globo(assunto,i,'oglobo') + get_links_from_google('oglobo.globo.com',assunto,num)
    links = [x for x in list(set(links)) if x not in oglobo_link]
    print('buscando em '+str(len(links))+' links em Oglobo')
    for link in links:
        oglobo = oglobo.append(get_news_oglobo(link.replace(' ','%20')), ignore_index=True)
    oglobo.to_csv('dados_oglobo190920.csv',index=False,sep='|',encoding='utf-8')
    return oglobo
```

```python
num = 3
shuffle(assuntos)
for assunto in assuntos:
    print(datetime.now(),assunto)
    for i in range(num+1):
        g1     = get_g1(assunto,i)
        r7     = get_r7(assunto,i)
        oglobo = get_oglobo(assunto,i)
        uol    = get_uol(assunto,i)
```

```python
df = pd.concat([oglobo,g1,uol,r7])
print('total de noticias ',df.shape[0])
```

```python
df['Desc_limpa'] = df['article'].astype(str).str.upper()
df['Desc_limpa'] = df['Desc_limpa'].apply(remover_acentos)
df['Desc_limpa'] = df['Desc_limpa'].str.replace('_',' ')
df['Desc_limpa'] = df['Desc_limpa'].str.replace('[^A-Za-z]',' ')
df['Desc_limpa'] = df['Desc_limpa'].str.strip()
```

```python
dic = df['Desc_limpa'].apply(lambda x: 
                             pd.value_counts([x for x in x.split(" ") if x not in  
                                                        (['', ' '] + get_stopwords())])).sum(axis = 0)
```

```python
dic = pd.DataFrame(dic,columns=['contagem']).sort_values('contagem',ascending=False)
dic.to_csv('dicionario190920.csv')
```
